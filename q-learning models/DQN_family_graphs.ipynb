```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
ALL_REWARDS = None
with open("./nash_dqn_experiments/models/reward_dict/my_dict.json") as f:
    ALL_REWARDS = json.load(f)
EPISODES = len(ALL_REWARDS['speaker_0'])
```

```python
EPISODES = 600
plt.plot(range(EPISODES), ALL_REWARDS['listener_0'][:EPISODES])
plt.title('Nash DQN: Avg reward per step for listener & speaker')
plt.xlabel('num_episodes')
plt.ylabel('reward')
plt.ylim(-13, 0.5)
plt.show()
```
